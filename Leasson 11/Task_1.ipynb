```python
from sympy import *
```

```python
x = Symbol("x")
func = -5 * x ** 2 + 10 * x + 150
y = solve(func)
x1 = round(float(y[0]), 2)
x2 = round(float(y[1]), 2)
print(f"x1 = {x1} x2 = {x2}")
```

```python
func_diff = diff(func)
print(solve(func_diff > 0))
print(solve(func_diff < 0))
```

```python
plotting.plot(func, (x, -5, 2.5))
```

```python
roots = solve(func_diff)
vertex = roots[0]
x = vertex
y = -5 * x ** 2 + 10 * x + 150
print(f"x = {vertex} y = {y}")
```
